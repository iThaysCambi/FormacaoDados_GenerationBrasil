```python
import pandas as pd

df_hotel = pd.read_csv ('/content/Limpeza-Dados-Hotel/hotel_dataset_messy.csv')
print('Informaçoes da base de dados')
df_hotel.info()
display(df_hotel.head(3))
```

```python
# transformando dados
# arrumando os nomes para modo de titulo
df_hotel['Guest_Name'] = df_hotel['Guest_Name'].str.title()

# fazendo datas serem datas e não objetos
df_hotel['CheckIn_Date'] = pd.to_datetime(df_hotel['CheckIn_Date'])
df_hotel['CheckOut_Date'] = pd.to_datetime(df_hotel['CheckOut_Date'])

# coluna nova chamada de tempo de hospedado - Days Hosted
df_hotel['Days Hosted'] = (df_hotel['CheckOut_Date']-df_hotel['CheckIn_Date']).dt.days

# arredondando os valores
df_hotel['Total_Charges'] = df_hotel['Total_Charges'].round(2)

# removendo duplicadas
df_hotel = df_hotel.drop_duplicates()
# removendo idades acima de 99 anos
df_hotel = df_hotel[df_hotel['Age'] <= 99]
#convertendo dados de negativo para positivo
df_hotel['Total_Charges'] = df_hotel['Total_Charges'].abs()
```

```python
display(df_hotel)
```

```python
#importando biblioteca de graficos
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
#configurar o estilo dos graficos
sns.set_theme(style='dark')
plt.figure(figsize=(15,5))

plt.subplot(1, 3, 1)
sns.histplot(df_hotel['Age'], bins=20, kde=True, color='blue')
plt.title('Distribution of Age')

plt.subplot(1, 3, 2)
sns.countplot(data=df_hotel, x='Satisfaction_Score',order=df_hotel['Satisfaction_Score'].value_counts().index, color= 'red')
plt.title('Distribution of Satisfaction Score')

plt.subplot(1, 3, 3)
sns.histplot(df_hotel['Days Hosted'], bins=10, kde=True, color='green')
plt.title('Distribution of Days Hosted')
plt.tight_layout()
plt.show()
```

### Interactive Visualizations with Plotly

Now, let's recreate the previous plots and add a new one for `Booking_Channel` using Plotly for interactive visualization.

```python
import plotly.express as px
import plotly.graph_objects as go

# Define a custom color palette for consistency or preference
custom_colors = px.colors.qualitative.Plotly # Or choose another palette like 'Pastel', 'Dark24', 'T10'

# 1. Distribution of Age
fig_age = px.histogram(df_hotel, x='Age', nbins=20, title='Distribution of Age',
                     color_discrete_sequence=[custom_colors[0]]) # Using a single color from the palette
fig_age.update_layout(bargap=0.1) # Add some gap between bars for better readability
fig_age.show()

# 2. Distribution of Satisfaction Score
satisfaction_counts = df_hotel['Satisfaction_Score'].value_counts().reset_index()
satisfaction_counts.columns = ['Satisfaction_Score', 'Count']
fig_satisfaction = px.bar(satisfaction_counts, x='Satisfaction_Score', y='Count',
                          title='Distribution of Satisfaction Score',
                          category_orders={'Satisfaction_Score': sorted(satisfaction_counts['Satisfaction_Score'].unique())},
                          color_discrete_sequence=[custom_colors[1]]) # Using another color
fig_satisfaction.show()

# 3. Distribution of Days Hosted
fig_days_hosted = px.histogram(df_hotel, x='Days Hosted', nbins=10, title='Distribution of Days Hosted',
                               color_discrete_sequence=[custom_colors[2]]) # Another color
fig_days_hosted.update_layout(bargap=0.1) # Add some gap between bars for better readability
fig_days_hosted.show()

# 4. Most Popular Booking Channel
booking_channel_counts = df_hotel['Booking_Channel'].value_counts().reset_index()
booking_channel_counts.columns = ['Booking_Channel', 'Count']
fig_booking_channel = px.bar(booking_channel_counts, x='Booking_Channel', y='Count',
                             title='Most Popular Booking Channels',
                             color='Booking_Channel', # Assign color by channel to differentiate
                             color_discrete_sequence=px.colors.qualitative.Set2) # A different palette for this plot
fig_booking_channel.show()
```
